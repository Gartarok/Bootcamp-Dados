1 Quanto cada cliente gastou em casa tipo de frete?

    SELECT 
        customer_id, 
        ship_via, SUM(freight) AS gasto_total
    FROM 
        orders
    GROUP BY 
        ship_via, customer_id
    ORDER BY 
        customer_id, ship_via;

2 O valor do produto mais barato, mais caro e a média dos valores unitários por fornecedor e categoria.  
| fornecedor | categoria | maior_valor | menor_valor | media_valores |

    SELECT 
        supplier_id, 
        category_id, 
        MAX(unit_price) AS maximo, 
        MIN(unit_price) AS minimo, 
        ROUND(AVG(unit_price)::numeric, 2) AS media
    FROM 
        products
    GROUP BY 
        supplier_id, category_id
    ORDER BY 
        supplier_id, category_id;
