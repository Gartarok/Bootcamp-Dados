### *2. Quantas ordens de compra há para cada categoria de produto?*

```python
#FINAL
```

    SELECT 
        products.category_id AS categoria,
        COUNT(order_details.order_id) AS total_ordens
    FROM 
        order_details
    INNER JOIN 
        products 
    ON 
        products.product_id = order_details.product_id
    GROUP BY
        products.category_id
    ORDER BY
        products.category_id ASC
