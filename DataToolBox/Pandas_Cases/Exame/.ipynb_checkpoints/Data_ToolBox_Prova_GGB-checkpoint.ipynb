## Questão 1

Considerando o conjunto de dados `natural-gas-prices-daily.csv`, carregue-o em um DataFrame e, em seguida:

a. (0,625) Visualize suas cinco primeiras linhas e imprima suas colunas. Em qual intervalo de datas temos os preços do gás natural?  
b. (0,625) Obtenha os preços mínimo e máximo do gás natural ao longo de todo o período de observação. Em quais datas o preço assumiu esses valores?  
c. (0,625) Crie um gráfico que te permita visualizar o preço do gás natural em cada dia, ao longo de todo o período de observação.  
d. (0,625) Crie uma nova coluna no DataFrame que contenha apenas o mês (como `int` ou a sigla como `str`) por meio da coluna `Date`.  
e. (0,625) Faça um gráfico que te permita visualizar o preço médio do gás natural ao longo dos 12 meses, considerando todo o período de observação (todos os meses).  
f. (0,625) Quais meses apresentam os preços mínimo e máximo do gás natural?

> **Valendo 3,75 pontos**

```python
#Importar o arquivo csv, chamando-o de df1
df1 = pd.read_csv('natural-gas-prices-daily.csv')
```

###### 1A
Visualize suas cinco primeiras linhas e imprima suas colunas. Em qual intervalo de datas temos os preços do gás natural?

```python
#Cinco primeras linhas de df1
df1.head(5)
```

```python
#colunas de df1
df1.info()
```

```python
#Converter o Date para datetime
df1['Date'] = pd.to_datetime(df1['Date'], format='%Y-%m-%d')
```

```python
#Intervalo de datas
print('O intervalo de tempo está entre as seguintes datas:')
print(df1['Date'].min())
print(df1['Date'].max())
```

###### 1B
Obtenha os preços mínimo e máximo do gás natural ao longo de todo o período de observação. Em quais datas o preço assumiu esses valores?

```python
minimo = df1['Price'].min()
maximo = df1['Price'].max()
print(f'Valor mínimo: {minimo} ')
print(f'Valor máximo: {maximo} ')

print('\nDia do mínimo')
print(df1['Date'][df1['Price'] == df1['Price'].min()])
print('\nDia do máximo')
print(df1['Date'][df1['Price'] == df1['Price'].max()])
```

##### 1C
Crie um gráfico que te permita visualizar o preço do gás natural em cada dia, ao longo de todo o período de observação.

```python
sns.lineplot(data=df1, x="Date", y="Price")
```

#### 1D
Crie uma nova coluna no DataFrame que contenha apenas o mês (como int ou a sigla como str) por meio da coluna Date.

```python
df1['Month']=df1['Date'].dt.month
df1
```

#### 1E
Faça um gráfico que te permita visualizar o preço médio do gás natural ao longo dos 12 meses, considerando todo o período de observação (todos os meses).

```python
media1 = df1.groupby(df1['Month']).mean()[['Price']]

sns.lineplot(data=media1)
plt.xlabel("Mês", size=10)
plt.ylabel("Preço médio", size=10)
plt.title("Mês x Preço médio", size=12)
```

##### 1F
Quais meses apresentam os preços mínimo e máximo do gás natural?

```python
index(media1).get_loc(media1.min())
```

```python
#Mês com o menor preço
print('Mês com o menor preço médio: ')
media1[media1['Price'] == media1['Price'].min()].index[0]
```

```python
#Mês com o maior preço
print('Mês com o maior preço médio: ')
media1[media1['Price'] == media1['Price'].max()].index[0]
```

## Questão 2

Com a tabela de vendas de jogos de vídeo-games (`videogames.csv`), indique:

a. (0,625) Quais jogos estão no Top 10 de Vendas Globais, considerando apenas aqueles pertencentes às plataformas PS3, Wii, X360 e PC.  
b. (0,625) A proporção (percentual) de jogos criados por gênero: tabela e gráfico de pizza.  
c. (0,625) Um gráfico de linha com a quantidade de vendas, em média, por ano, separadas pelas plataformas PS3, Wii, X360 e PSP.  

> **Valendo 1,875 ponto**

```python
#Importar o arquivo csv, chamando-o de df2
df2 = pd.read_csv('videogame.csv')
df2.info()
```

#### 2A
Quais jogos estão no Top 10 de Vendas Globais, considerando apenas aqueles pertencentes às plataformas PS3, Wii, X360 e PC.

```python
df2
```

```python
#Top 10 para Wii
df2.where(df2['Platform'] == 'Wii').sort_values(by=['Global_Sales'], inplace=False, ascending=False).head(10)
```

```python
#Top 10 para PS3
df2.where(df2['Platform'] == 'PS3').sort_values(by=['Global_Sales'], inplace=False, ascending=False).head(10)
```

```python
#Top 10 para X360
df2.where(df2['Platform'] == 'X360').sort_values(by=['Global_Sales'], inplace=False, ascending=False).head(10)
```

```python
#Top 10 para PC
df2.where(df2['Platform'] == 'PC').sort_values(by=['Global_Sales'], inplace=False, ascending=False).head(10)
```

#### 2B
A proporção (percentual) de jogos criados por gênero: tabela e gráfico de pizza.

```python
df2.info()
```

```python
#Tabela, percentual de jogos
data2 = df2['Genre'].value_counts(normalize=True)
data2
```

```python
#O famoso gráfico de comida
colors = sns.color_palette('pastel')[0:5]
plt.pie(data2, labels = data2.index, colors = colors, autopct='%.1f%%', normalize=False)
plt.show()
```

#### 2C
Um gráfico de linha com a quantidade de vendas globais, em média, por ano, separadas pelas plataformas PS3, Wii, X360 e PSP.

```python
# Plotagem das linhas (gráficos)
lista = ['PS3','Wii','X360','PSP']
for i in lista:
    group = df2[df2['Platform'] == i]
    group2= group.groupby(['Year']).mean()[['Global_Sales']]
    plt.plot(group2.index, group2, label=i)

# Recursos de estilização
plt.xlabel('Ano')
plt.ylabel('Média de vendas globais')
plt.grid()
plt.legend()
```

```python
PS3, Wii, X360, PSP = 0,0,0,0
dfgrafico = pd.DataFrame()
lista = ['PS3','Wii','X360','PSP']
for i in lista:
    group = df2[df2['Platform'] == i]
    group2= group.groupby(['Year']).mean()[['Global_Sales']]
    #dfgrafico = dfgrafico.append(group2, ignore_index = True)
    pd.concat([group2, group2], axis=1)
dfgrafico
```

## Questão 3

Considerando os dados do dataset `fish.csv`, faça uma análise exploratória dos dados, criando:

a. (0,625) Vários gráficos de dispersão (em uma mesma figura) combinando cada par de característica (*feature*) dos peixes, e atribuindo uma cor diferente para cada espécie.  
b. (0,625) Crie um gráfico de distribuição da largura dos peixes (`Width`). Não deixe de incluir o KDE.  
c. (0,625) Crie um mapa de calor que apresente a correlação entre cada *feature* do peixe. Anote os dados no gráfico e utilize o `cmap=coolwarm`.  

> **Valendo 1,875 ponto**

```python
#Importar o arquivo csv, chamando-o de df3
df3 = pd.read_csv('fish.csv')
df3
```

```python
df3.info()
```

```python
#Peixes unicos
peixes=list(pd.unique(df3['Species']))
peixes
```

#### 3A
Vários gráficos de dispersão (em uma mesma figura) combinando cada par de característica (feature) dos peixes, e atribuindo uma cor diferente para cada espécie.

```python
#ABORDAGEM 
#Usar o pairplot para fazer tudo, usando como tom as especies
#diag_kind = None, porque o gráfico padrão da diagonal é de densidade e não dispersão
sns.pairplot(df3, hue='Species', kind='scatter', diag_kind=None)
```

#### 3B
Crie um gráfico de distribuição da largura dos peixes (`Width`). Não deixe de incluir o KDE.  


```python
#Entendi que é a distruição geral dos peixes aqui
sns.histplot(df3['Width'], kde=True)
```

```python
#Com todos os peixes
sns.histplot(data = df3, x='Width', hue='Species',kde=True)
```

#### 3C
Crie um mapa de calor que apresente a correlação entre cada *feature* do peixe. Anote os dados no gráfico e utilize o `cmap=coolwarm`.  


```python
#Mapa de calor com as correlações [df3.corr()]
sns.heatmap(df3.corr(), cmap='coolwarm', vmin=0, vmax=1, annot=True)
```

```python
#Importar o arquivo csv, chamando-o de df4
df4 = pd.read_csv('crimes.csv',sep='|')
df4.shape
```

```python
#Gráficos
grid = sns.FacetGrid(df4, col="periodo",hue="Category")
grid.map_dataframe(sns.scatterplot, x="X", y='Y')
grid.set_axis_labels("Latitude", "Longitude")
grid.add_legend()
```
