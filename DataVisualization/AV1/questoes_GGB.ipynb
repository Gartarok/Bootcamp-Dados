### A. Linhas  
Os gráficos de linha usam pontos conectados para demonstrar a variação de um valor. O eixo horizontal expressa uma progressão continua, usualmente sendo o tempo, e o vertical o valor que se quer expressar a variação conforme o horizontal.

1. Relacionar a velocidade de um objeto com a variação do tempo. Com o gráfico de linha podemos observar a variação acentuada no ângulo das linhas e inferir que este está, por exemplo, com uma aceleração posisitva.



2. Analisar mais de um parametro conforme a variação do tempo. Podemos por exemplo analisar a variação da precipitação e variação da temperatura conforme variam os meses do ano, e com isso podemos fazer conclusões sobre a variação das caracteristicas dos meses do ano.



### B. Barras  
Grafícos de barra são utilizados para expressar o valor total de alguma catégoria, comparando com outras categorias ou 

1. Uma boa aplicação é em gráficos de vendas, pois podemos visualizar facilmente qual a marca ou categoria que mais vende e qual o valor.



2. É uma prática também utiliza-lo para comparar as vendas ou incidencia de multiplos evento em situações diferentes, como ao longo de anos ou conforme a sazionalidade. Pois, o gráfico de barras deixa essa vizualização simples devido a sua estrutura.



  
### C. Texto Simples
  
Uma visualização simples que informa de uma forma muito direta valores ou variações em determinado parâmetro

1. Anuncio de variação de preços. Pois, apresenta fatos de forma simples e direta sobre produtos importantes como combustíveis.



2. Indicação da alteração de um parametro importante, como temperatura global, taxas, indicadores economicos e afins. Pois, são valores muito objetivos e que podem ser resumidos a apenas ao valor atual, em alguns casos.



### D. Mapa de Calor (heatmap)


O mapa de calor (também conhecido como mapa de calor) mostra valores para uma variável principal de interesse em duas variáveis ​​de eixo como uma grade de quadrados coloridos. As variáveis ​​do eixo são divididas em intervalos como um gráfico de barras ou histograma, e a cor de cada célula indica o valor da variável principal no intervalo de células correspondente.

O mapa de calor mostra o valor que relaciona as duas váriaveis dispostas nos eixos. Por exemplo, para um determinado dia em um determinado Mês mostra a temperatura média ao longo do dia.
1. Para indicar a temperatura de várias cidade ao longo de um périodo de tempo. O mapa de calor é ideal para isso, pois consegue deixar claro a temperatura conforme variam os paramêtros (cidade e mês-ano).


2. Ele também pode ser usado muito bem para identificar através de cor a densidade de atividades, mesmo sem aprentar o valor a primeira vista. Como por exemplo no gráfico de atividade do Github.



E. Gráfico de dispersão (scatterplot)
Gráficos de dispersão usam pontos para representar o valor para duas variaveis númericas diferentes. Esse tipo de gráfico é usado para observar a relação entre duas variáveis.

1. Scatter plots são usados para indicar que as variáveis envolvidas tem relação, pois dependendo da forma que fica o gráfico final podemos ver essa relação. Por exemplo, se quando uma váriavel aumenta e a outra também podemos inferir que elas estão intimamente ligadas, mas se elas apresentam um compartamente aleatório em relação a outra, logo não podemos concluir nada.



2. Casos reais onde queremos saber a relação entre as varíaveis e temos uma escala tão grande em um dos eixos que para ficar adequado deve estar em escala logaritmica. O scatter plot fica adequado, pois é possível visualizar a tendência no caso do uso de log em um dos eixos.
