```python
#Eliminar o Balance porque é o que queremos
#Eliminar o Limit pois é muito com o balance, acima de 85%
X1 = df_numerico.drop(columns=['Balance','Limit'])
Y1 = df_numerico['Balance']
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1,Y1,test_size=0.3,random_state=2)
```

```python
lr1 = LinearRegression().fit(X1,Y1)
```

```python
cross_validate(estimator=LinearRegression(),X=X_train1,y=y_train1,scoring='r2',
              return_train_score=True)
```

```python
lr1 = LinearRegression().fit(X_train1, y_train1)
```

```python
y_pred1 = lr1.predict(X_test1)
```

```python
y_pred1[:5]
```

```python
y_test1[:5]
```

```python
resids1 = y_test1 - y_pred1
resids1
```

```python
def plot_resids(resids, y_true):
    plt.figure(figsize=(18,7))
    
    plt.subplot(1,2,1)
    plt.scatter(y_true, resids)
    
    plt.subplot(1,2,2)
    sns.histplot(resids, kde=True)
```

```python
# Visivel um outlier, acima de 300, eixo Y
# Muitos zeros
plot_resids(resids1,y_test1)
```

```python
#Novo modelo, tratando apenas os ativos
```

```python
df_dummies = pd.get_dummies(df,drop_first=True)
df_dummies.head()
```

```python
df_dummies['Active']=df['Balance'].map(lambda balance: 0 if balance ==0 else 1)
df_dummies.head()
```

```python
X2 = df_dummies.drop(columns=['Balance'])
y2 = df_dummies['Balance']
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2,y2,test_size=0.3,random_state=2)
```

```python
cross_validate(estimator=LinearRegression(),X=X_train2,y=y_train2,scoring='r2',
              return_train_score=True)
```

```python
lr2=LinearRegression().fit(X_train2,y_train2)
y_pred2=lr2.predict(X_test2)
resids2=y_test2-y_pred2
```

```python
plot_resids(resids2,y_test2)
```

```python
y_pred2[:5]
```

```python
y_test2[:5].values
```

### Considerando apenas ativos, Balance !=0

```python
X_active = df_dummies[df_dummies['Active']==1]
X_active.shape
```

```python
y_active = df_dummies[df_dummies['Balance']>0]['Balance']
y_active.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_active,y_active,random_state=2)
```

```python
lr=LinearRegression().fit(X_train,y_train)
y_pred = lr.predict(X_test)
resids = y_test - y_pred
```

```python
plot_resids(resids=resids,y_true=y_test)
```
